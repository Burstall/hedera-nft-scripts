```python
import pandas as pd
pd.options.display.float_format = '{:.0f}'.format
tokenId = '_metadata.json-2022-09-21T06-18-25'
```

```python
metadata_df = pd.read_table(tokenId+'.tsv')
metadata_df.head()
```

```python
metadata_df.info()
metadata_df.columns.values.tolist()
```

```python
column_list = [
 'figure',
 'shoes',
 'topping',
 'hat',
 'eyes',
 'lefthand',
 'righthand',
 'feels'
]
```

```python

for column_name in column_list:
    df_norm = metadata_df[column_name].value_counts(ascending=True, normalize=True, dropna=False)
    df_count = metadata_df[column_name].value_counts(ascending=True, dropna=False)
    df_stat_table = pd.concat([df_norm, df_count], axis=1, join='inner')
    df_stat_table.set_axis(["percentage", "count"], axis=1, inplace = True)
    df_stat_table.loc['Column_Total'] = df_stat_table.sum(numeric_only=True, axis=0)
    df_stat_table['percentage'] = pd.Series(["{0:.2f}%".format(val * 100) for val in df_stat_table['percentage']], index = df_stat_table.index)
    print(df_stat_table)
```

```python
#produce the stats tables
output_str = ""
for column_name in column_list:
    df_norm = metadata_df[column_name].value_counts(ascending=True, normalize=True, dropna=False)
    df_count = metadata_df[column_name].value_counts(ascending=True, dropna=False)
    df_stat_table = pd.concat([df_norm, df_count], axis=1, join='inner')
    df_stat_table.set_axis(["percentage", "count"], axis=1, inplace = True)
    df_stat_table.loc['Column_Total'] = df_stat_table.sum(numeric_only=True, axis=0)
    df_stat_table['percentage'] = pd.Series(["{0:.2f}%".format(val * 100) for val in df_stat_table['percentage']], index = df_stat_table.index)
    output_str = output_str + "\n\n" + column_name + ":\n" + df_stat_table.to_string()

output_str
#with open("0.0.848553_stats.txt", "w") as a_file:
#   a_file.write(output_str)
```

```python
#process the data for % and points then sort and rank
column_pts_lst = []

for column_name in column_list:
    metadata_df[column_name +'_perc'] = metadata_df[column_name].map(metadata_df[column_name].value_counts(normalize=True, dropna=False))
    metadata_df[column_name +'_pts'] = 1/metadata_df[column_name +'_perc']
    column_pts_lst.append(column_name +'_pts')

#column_pts_lst.append('total')
metadata_df['total_pts'] = metadata_df[column_pts_lst].sum(axis=1)

sorted_metadata_df = metadata_df.sort_values(by=['total_pts','serial'], ascending=(False,True)).reset_index()
sorted_metadata_df['rank'] = sorted_metadata_df.index + 1
sorted_metadata_df.drop('index', axis=1, inplace=True)
```

```python
sorted_metadata_df.to_csv(tokenId+'_processed.tsv', index=False, sep = '\t')
sorted_metadata_df.to_csv(tokenId+'_processed.csv', index=False)

with open(tokenId+'_stats.txt', "w") as a_file:
   a_file.write(output_str)
```
